```python
Employee=[["Employee ID","Employee Name","Salary","Department Number"]]
E_Id=[]

def Add_emp():
    
    Empid=(int(input("Employee ID '(3 Digit number)':")))
    while(Empid<=99):
        Empid=int(input("Enter a Valid 3 Digit number:"))
    while(Empid in E_Id):
        Empid=int(input("Duplicate value-Enter a Unique Employee-ID:"))
    E_Id.append(Empid)
    
    
    Ename=input("Employee Name:").upper()
    Sal=int(input("Salary'(Above 3000)':"))
    while(Sal<3000):
        Sal=int(input("Enter a Valid Salary '(Above 3000)':"))
        
    Deptno=int(input("Department Number 'Valid Numbers (10,20,30)':"))
    while(Deptno not in [10,20,30]):
        Deptno=int(input("Enter a Dept Number '(Must be - 10 (or) 20 (or) 30)':"))
    Employee.append([E_Id[-1],Ename,Sal,Deptno])
    return
        
def Display_emp():
    f=open('emp.txt',"w")
    for i in Employee:
        for j in range(len(i)):
            if(j==len(i)-1):
                f.write(str(i[j])+"\n")
            else:
                f.write(str(i[j])+",")
    f.close()
    f=open('emp.txt','r')
    k=f.readlines()
    for i in (k):
        print(i)
    return
    
    
def Seperate_emp():
    for i in Employee[1:]:
        if(i[-1]==10):
            f=open('Emp_10.txt',"w")
            for j in range(len(i)):
                if(j==len(i)-1):
                    f.write(str(i[j])+"\n")
                else:
                    f.write(str(i[j])+",")   
        elif(i[-1]==20):
            f=open('Emp_20.txt',"w")
            for j in range(len(i)):
                if(j==len(i)-1):
                    f.write(str(i[j])+"\n")
                else:
                    f.write(str(i[j])+",")
        elif(i[-1]==30):
            f=open('Emp_30.txt',"w")
            for j in range(len(i)):
                if(j==len(i)-1):
                    f.write(str(i[j])+"\n")
                else:
                    f.write(str(i[j])+",")
    
    
while(1):
    x=int(input("1.Add_emp() \n2.Display_emp() \n3.Seperate_emp \n4.Exit \n"))
    if(x==1):Add_emp()
    elif(x==2):Display_emp()
    elif(x==3):Seperate_emp()
    elif(x==4):
        print('Session Exited')
        break
    else:print("\nEnter a Valid Choice: ")
```
